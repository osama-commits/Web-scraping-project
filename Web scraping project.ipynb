```python
years = [2022]

fill_url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for year in years:
    url = fill_url.format(year)
    print(url)
```

```python
import pandas as pd
```

```python
df = pd.read_html(url, header = 0) # to define the first row as the header of the dataframe
df
```

```python
# to know how many tables are in the dataframe
len(df)
```

```python
# to select the first table 
df[0]
```

```python
# assigning the table to a variable
df2022 = df[0]
```

Data cleaning

```python
# find the extra headers in the table 
df2022[df2022.Age == "Age"]
```

```python
# remove the extra headers
df_extra_headers_dropped = df2022.drop(df2022[df2022.Age == "Age"].index)
df2022.shape
```

```python
df_extra_headers_dropped.shape
```

Data visualization

```python
import seaborn as sns
```

plot a histogram of all the players' score

```python
# kde is set to false to keep the y axis points according to the players' points
sns.distplot(df_extra_headers_dropped.PTS, kde = False)
```

Add bar line colour and edgewidth

```python
sns.distplot(df_extra_headers_dropped.PTS, kde = False, hist_kws = dict(edgecolor = "blue", linewidth = 2))
```

```python
sns.distplot(df_extra_headers_dropped.PTS, kde = False, hist_kws = dict(edgecolor = "black", linewidth = 2), color = "gold")
```
